   - the above list in decreasing order
   - all values greater than 15

```python
def lists(lst):
    '''
    Return three lists with different conditions
    '''
    a = list(lst)
    b = list(lst)
    c = list(lst)
    a = [1 if item == 6 else item for item in lst]
    b.sort()
    b.reverse()
    c = [item for item in lst if item > 15]
    return a,b,c
# Test
lst =  [23, 54, 6, 8, 10078, 3]
lists(lst)
    
```
